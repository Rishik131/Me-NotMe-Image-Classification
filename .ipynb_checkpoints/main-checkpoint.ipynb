### Viewing Images

```python
img = cv2.imread(current_dir+'/dataset/20.png')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.xticks([])
plt.yticks([])
```

### Pre-Processing

```python
shutil.move('./dataset/20.png','./dataset/me/20.png')
```

```python
for file in files:
    if file == '20.png': continue
    shutil.move('./dataset/'+file,'./dataset/notme/'+file)
```

```python
current_dir = os.getcwd()+'/dataset/me'
current_dir
```

```python
augments = aug.Affine(translate_percent={'x':(-0.2,0.2),'y':(-0.2,0.2)},rotate=(-30,30),scale=(0.5,1.5))
```

```python
imgs = []
for i in range(60):
    imgs.append(img)
```

```python
aug_imgs = augments(images=imgs)
```

```python
fig,ax = plt.subplots(10,6)
i,j=0,0
for img1 in aug_imgs:
    ax[i,j].imshow(img1)
    ax[i,j].set_xticks([])
    ax[i,j].set_yticks([])
    if j==5:
        i+=1
        j=0
    else: j+=1
```

```python
me = []
me.append(img)
for img1 in aug_imgs:
    me.append(img1)
```
